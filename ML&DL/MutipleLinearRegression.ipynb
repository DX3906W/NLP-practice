```
import numpy as np
import tensorflow as  tf
```

```
boston = tf.contrib.learn.datasets.load_dataset('boston')
m, n = boston.data.shape
print(boston.data.shape)

```

```

learning_rate = 0.01
epoches = 50

train_x = boston.data.astype(np.float32)
train_y = boston.target.reshape(-1, 1).astype(np.float32)
print(train_y.shape)

X = tf.placeholder(dtype=tf.float32, shape=[None, n], name='X')
y_ = tf.placeholder(dtype=tf.float32, shape=[None, 1], name='y_')

W = tf.Variable(tf.random_normal([n, 1], stddev=2), dtype=tf.float32, name='weight')
b = tf.Variable(tf.ones([1], dtype=tf.float32, name='bias'))

loss_function = tf.reduce_mean(tf.square(y_ - tf.add(tf.matmul(X, W),b)))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss_function)

init = tf.global_variables_initializer()

with tf.Session() as sess:
  sess.run(init)
  for epoch in range(epoches):
    for xs, ys in zip(train_x, train_y):
      xs = xs.reshape(n, )[None,:]
      ys = ys.reshape(1, 1)
      #print(xs.dtype)
      # print(ys)
      sess.run(optimizer, feed_dict={X: xs, y_: ys})
    if epoch%10==0:
      print(sess.run(loss_function))
  W_ = sess.run(W)
  print(W_)
```
