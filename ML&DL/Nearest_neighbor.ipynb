```
%tensorflow_version 1.x
import tensorflow as tf
import numpy as np

from tensorflow.examples.tutorials.mnist import input_data
```

```
mnist = input_data.read_data_sets('/tmp/data',  one_hot=True)

train_x, Y = mnist.train.next_batch(5000)
test_x, Y_ = mnist.train.next_batch(200)
#print(Y)

X = tf.placeholder(dtype=tf.float32, shape=[None, 784], name='train_x')
X_ = tf.placeholder(dtype=tf.float32, shape=[784], name='test_x')

distance = tf.reduce_sum(tf.abs(tf.add(X, tf.negative(X_))), reduction_indices=1)

pred = tf.arg_min(distance, 0)
accuracy = 0

init = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
  sess.run(init)
  for i in range(len(test_x)):
    #得到5000个训练样本对于该测试样本的预测结果
    nn_index = sess.run(pred, feed_dict={X:train_x, X_:test_x[i, :]})
    print("Test", i, "Prediction:", np.argmax(Y[nn_index]), \
            "True Class:", np.argmax(Y_[i]))
    if np.argmax(Y[nn_index])==np.argmax(Y_[i]):
      accuracy += 1./len(test_x)
  print('done!')
  print('Accuracy', accuracy)
```
